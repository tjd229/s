## 4.2 조건부 확률

- 주머니에서 카드를 뽑아 상자를 선택하고 선택된 상자에서 공을 뽑는다
<img src="images/boxandball.png"/><br/>
- 확률 P(A) = 상자 A가 선택될 확률 = 7/10
- P(A)를 사전확률이라고 함
- 확률 P(하양|A) = 상자 A에서 하얀공이 나올 확률 = 2/10
- P(하양|A)를 조건부확률이라 함.
    - P(S|W) = W가 일어났을 때, S일 확률
- 확률 P(하양∩A) = 상자 A가 뽑히고, 하얀공이 나올 확률 = 7/10 x 2/10 = 7/50
    - P(하양∩A)=P(A)P(하양|A)


- 하얀 공이 뽑혔는데 어느 상자에서 나왔는지 알 수 있는 방법은?
- 기본 전략: 상자 A와 B에서 나왔을 가능성을 각각 구하고 큰 가능성을 보인 상자를 답으로 취한다.
- 방법1 : P(하양|B), P(하양|A) 각 상자에서의 하얀 공이 나타날 확률 인 조건부 확률을 사용하여 구함
    - P(하양|B) 9/15 > P(하양|A) = 2/10 이므로 “상자 B에서 나왔다.” 
    - 이러한 조건부 확률을 우도(Likelihood)라 부름
- 방법2 : 사전 확률을 사용하여 상자 A와 상자 B의 선택 가능성을 비교
    - P(A) = 7/10 > P(B) = 3/10 이므로 “상자 A에서 나왔다.”

- 두 가지 방법의 한계
    - 극단적으로 P(A) = 0.999 라면 방법 1이 틀린 것이 확실하다.
    - 극단적으로 P(하양|A) = 0.999 라면 방법2가 틀린 것이 확실하다.
    - 우도(조건부 확률)와 사전 확률을 모두 고려해야 함
- 해결 방법
    - 조건부 확률 P(A|하양)과 P(B|하양)을 비교하여 큰 쪽을 취함
    - P(A|하양), P(B|하양)을 사후 확률이라고 함

## 4.3 조건부 확률로 분류하기

- A라는 사건이 일어났을 때, 분류를 c1으로 할것이냐 c2로 할것이냐를 식으로 나타내면
    - P(c1|A) or P(c2|A)

- 베이스 규칙을 가지고 다음과 같이 정의 할 수 있음
    - P(c1|A) > P(c2|A)이면, 분류 항목 c1에 속한다.
    - P(c1|A) < P(c2|A)이면, 분류 항목 c2에 속한다.

## 4.5 파이썬으로 텍스트 분류하기
### 4.5.1 준비: 텍스트로 단어 벡터 만들기

```python
from numpy import *
def loadDataSet():
    postingList=[['my','dog','has','flea','problems','help','please'],
                 ['maybe','not','take','him','to','dog','park','stupid'],
                 ['my','dalmation','is','so','cute','I','love','him'],
                 ['stop','posting','stupid','worthless','garbage'],
                 ['mr','licks','ate','my','steak','how','to','stop','him'],
                 ['quit','buying','worthless','dog','food','stupid']]
    classVec=[0,1,0,1,0,1]  #1: 폭력적인, 0: 폭력적이지 않음
    return postingList,classVec

def createVocabList(dataSet):
    vocabSet=set([])
    for document in dataSet:
        vocabSet=vocabSet|set(document) 
    return list(vocabSet)

def setOfWords2Vec(vocabList,inputSet):
    returnVec=[0]*len(vocabList)
    for word in inputSet:
        if word in vocabList:
            returnVec[vocabList.index(word)]=1
        else: print "the word: %s is not in my Vocabulary!" % word
    return returnVec
```

- loadDataSet() : 단어 리스트와 그 리스트가 폭력적인지 아닌지 나타내는 분류값을 반환
    - 반환 값인 postingList는 배열의 배열 즉 matrix형태<br/><br/>
- createVocabList(dataSet) : 단어 리스트 dataSet중 중복된 것을 제거하여 리스트로 반환, 반환값은 아래 setOfWords2Vec에서 사용<br/>
    - vocabSet=vocabSet|set(document) : document가 vocabSet에 없으면 추가, 있으면 추가안하는 or 연산
    <br/><br/>
- setOfWords2Vec(vocabList, inputSet) : inputSet의 단어 들이, vocabList에 있으면 1, 없으면 0인 벡터리스트 반환
    - returnVec=[0]*len(vocabList) : 값이 0인 배열을 vocabList의 길이만큼 만듦

### 4.5.2 훈련: 단어 벡터로 확률 계산하기

- Pseudo Code
<img src="images/pseudo.png"/>

```python
def trainNB00(trainMatrix,trainCategory):
    numTrainDocs=len(trainMatrix)
    numWords=len(trainMatrix[0])  ##각 분류 항목에 대한 문서의 개수 세기
    pAbusive=sum(trainCategory)/float(numTrainDocs)  ##사전확률
    p0Num=zeros(numWords);p1Num=zeros(numWords)
    p0Denom=0.0;p1Denom=0.0
    for i in range(numTrainDocs):  ##훈련을 위한 모든 문서의 개수만큼 반복
        if trainCategory[i]==1:    ##분류 항목 개수만큼 반복
            p1Num+=trainMatrix[i]  ##해당 토큰이 문서 내에 있다면 -> 해당 토큰에 대한 개수를 증가
            p1Denom+=sum(trainMatrix[i]) ##토큰에 대한 개수를 증가
        else:
            p0Num+=trainMatrix[i]
            p0Denom+=sum(trainMatrix[i])
    p1Vect=(p1Num/p1Denom )  ##조건부 확률을 구하기 위해 해당 토큰의 개수를 전체의 수로 나눔
    p0Vect=(p0Num/p0Denom)
    return p0Vect,p1Vect,pAbusive #간 분류 항목에 대한 조건부 활률을 반환함
```

- trainNB00(trainMatrix,trainCategory)
    - trainMatrix : 단어 벡터의 matrix
    - trainCategory : matrix의 한 행당 해당되는 분류값
    - pAbusive : 폭력적인 문서가 있는 확률
    - zeros(n) : 길이가 n인 0으로 이루어진 배열 생성(numpy)
    - p1Vect : 분류가 1인 문서에서 해당 단어가 나올 확률의 벡터 리스트
    - p0Vect : 분류가 인 문서에서 해당 단어가 나올 확률의 벡터 리스트

### 4.5.3 검사: 실제 조건을 반영하기 위해 분류기 수정하기

```python
def trainNB0(trainMatrix,trainCategory):
    numTrainDocs=len(trainMatrix)
    numWords=len(trainMatrix[0])
    pAbusive=sum(trainCategory)/float(numTrainDocs)
    p0Num=ones(numWords);p1Num=ones(numWords)  ##zeros->ones
    p0Denom=2.0;p1Denom=2.0  ##0.0->2.0
    for i in range(numTrainDocs):
        if trainCategory[i]==1:
            p1Num+=trainMatrix[i]
            p1Denom+=sum(trainMatrix[i])
        else:
            p0Num+=trainMatrix[i]
            p0Denom+=sum(trainMatrix[i])
    p1Vect=log(p1Num/p1Denom)      ## p1Num/p1Denom -> log(p1Num/p1Denom)
    p0Vect=log(p0Num/p0Denom)      ## p0Num/p0Denom -> log(p0Num/p0Denom)
    return p0Vect,p1Vect,pAbusive

```

- trainNB0
    - 확률 벡터에 log를 취하도록 변경
    - 이전 함수의 결함을 처리하기 위한 변경
        - 1_조건부확률에서 곱셈연산시 0을 곱하여, 전체 곱이 0이 되는 것을 방지
        - 2_확률값이 아주 작을 경우 곱셈시 언더플로우가 일어나는 것을 방지
    - log를 취해도 아래와 같이 분포모양은 비슷하며, 나이브 베이시안 분류기는 확률의 상대적인 크고작음으로 분류하는 것이지 값을 가지고 분류하는게 아님
        - 음수가 나오는 이유 : 확률이 0~1사이의 값이므로 10**m(m<0)의 형태를 가지기 때문

```python
import matplotlib.pyplot as plt
%matplotlib inline

def func(x):
    return -(x-0.25)**2+1

a=linspace(0,0.5,200)
plt.plot(a,[func(x) for x in a])
plt.show()
plt.plot(a,[log(func(x)) for x in a])
plt.show()
```

```python
def classifyNB(vec2Classify,p0Vec,p1Vec,pClass1):
    p1=sum(vec2Classify*p1Vec)+log(pClass1)
    p0=sum(vec2Classify*p0Vec)+log(1.0-pClass1)
    if p1>p0:
        return 1
    else:
        return 0
    
def testingNB():
    ##훈련 과정
    listOPosts,listClasses=loadDataSet()
    myVocabList=createVocabList(listOPosts)
    trainMat=[]
    for postinDoc in listOPosts:
        trainMat.append(setOfWords2Vec(myVocabList,postinDoc))
    p0V,p1V,pAb=trainNB0(array(trainMat),array(listClasses))
    ####
    ####분류 테스트
    
    testEntry=['love','my','dalmation']
    thisDoc=array(setOfWords2Vec(myVocabList,testEntry))
    print testEntry, 'classfied as: ', classifyNB(thisDoc,p0V,p1V,pAb)
    testEntry=['stupid','garbage']
    thisDoc=array(setOfWords2Vec(myVocabList,testEntry))
    print testEntry, 'classfied as: ', classifyNB(thisDoc,p0V,p1V,pAb)
```

- classifyNB(vec2Classify,p0Vec,p1Vec,pClass1)
    - vec2Classify : 분류할 리스트
    - p0Vec,p1Vec : log 확률 벡터 리스트
    - pClass1 : 문서가 1일 확률
    - 더 높은 확률을 가진 문서 class를 반환

### 4.5.4 준비: 중복 단어 문서 모델

```python
def bagOfWords2VecMN(vocabList,inputSet):
    returnVec=[0]*len(vocabList)
    for word in inputSet:
        if word in vocabList:
            returnVec[vocabList.index(word)]+=1
    return returnVec
```

- 해당 단어가 있냐 없냐로 표현하는 것을 count로 변경
- 많이 나오는 단어는 1이 아닌 나온 수만큼 값이 할당되고, 그만큼 가중치도 오르게 됨

```python
def calcMostFreq(vocabList,fullText):
    import operator
    freqDict={}
    for token in vocabList:
        freqDict[token]=fullText.count(token)
    sortedFreq=sorted(freqDict.iteritems(),key=operator.itemgetter(1),reverse=True)
    return sortedFreq[:30]
    
def localWords(feed1,feed0):
    import feedparser
    #### 훈련 데이터 입력
    w1='entries'
    w2='summary'
    docList=[]; classList=[]; fullText=[]
    minLen=min(len(feed1[w1]),len(feed0[w1]))
    for i in range(minLen):
        wordList=textParse(feed1[w1][i][w2])
        docList.append(wordList)
        fullText.extend(wordList)
        classList.append(1)
        wordList=textParse(feed0[w1][i][w2])
        docList.append(wordList)
        fullText.extend(wordList)
        classList.append(0)
    vocabList=createVocabList(docList)
    #### 훈련 데이터 입력
    
    #### 접속사와 같은 빈도는 높지만 의미없는 단어 삭제를 위한 top30Words목록 생성
    top30Words=calcMostFreq(vocabList,fullText)
    for pairW in top30Words:
        if pairW[0] in vocabList:vocabList.remove(pairW[0])
    #### 접속사와 같은 빈도는 높지만 의미없는 단어 삭제를 위한 top30Words목록 생성\
    
    #### 더 짧은 feed의 길이의 두배 만큼의 범위에서 테스트 집합 추출
    #### 인덱스의 out of bound 방지
    trainingSet=range(2*minLen); testSet=[]
    for i in range(20):
        randIndex=int(random.uniform(0,len(trainingSet)))
        testSet.append(trainingSet[randIndex])
        del(trainingSet[randIndex])
    trainMat=[]; trainClasses=[]
    for docIndex in trainingSet:
        trainMat.append(bagOfWords2VecMN(vocabList,docList[docIndex]))
        trainClasses.append(classList[docIndex])
    #### 테스트 집합 추출 완료
    p0V,p1V,pSpam=trainNB0(array(trainMat),array(trainClasses))  ####분류기 생성
    errorCount=0
    for docIndex in testSet:
        wordVector = bagOfWords2VecMN(vocabList,docList[docIndex])
        if classifyNB(array(wordVector),p0V,p1V,pSpam)!=classList[docIndex]:
           errorCount+=1
    print 'the error rate is: ', float(errorCount)/len(testSet)
    return vocabList,p0V,p1V
```

## 4.9 베이지언 분류기의 확장과 한계

- 에러 확률 분포(사전 확률이 같은 경우)
    - 베이지언 분류기의 확률 밀도 그래프
<img src="images/2_5.jpg"/><br/>
    - 파란색 영역은 에러 확률을 나타냄
    - 중간 부분의 경우, 확률차이가 크게 나지 않기 때문에, 신빙성이 떨어짐
    - 그런 경우, 기각처리를 하여, 오류 빈도를 낮추는 방법이 있음
- 최소 위험 분류기
    - 암환자를 정상인으로 잘못 분류하는 것과 정상인을 암환자로 잘못 분류하는 것의 손실은 같지 않다.
    - 잘못 분류시의 손실을 각각 따져 베이지언 분류기를 수정하여 최소 위험 분류기가 만들어짐
    - w1=암환자, w2=정상인, c12=암환자를 정상인으로 잘못분류했을 때 리스크, c21=정상인을 암환자로 잘못 분류했을 때 리스크
    - c12*p(x|w1) > c21*p(x|w2)이면 w1로 분류하고,
    - c12*p(x|w1) < c21*p(x|w2)이면 w2로 분류
<img src="images/2_7.jpg"/><br/>
- 본 예제에서의 한계
    - 각 확률이 독립적인 확률이라 가정(실제로는 종속적인 케이스가 많음)
        - 성능 저하 고려
    - 분류를 2가지 상황에서만 적용
        - M 부류 베이지언 분류기
    - p(w|x)는 x가 일어났을 때 w일 확률인데, 연쇄적인 상황에서는 적용할 수 없음
        - 베이지언 네트워크
